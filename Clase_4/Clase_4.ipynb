Dado que la funcion de posicion esta dada como: Asen(wt), donde w = sqrt(k/m). 
La derivada teorica estaria dada por: Awcos(wt) 

```python
def dx_exacta(t):
    w = num.sqrt(400)
    return(2*w*num.cos(w*t))

def dx_atras(f,t):
    return(f(2,800,2,t)-f(2,800,2,t-epsi))/epsi

def dx_central(f,t):
    return(f(2,800,2,t+epsi)-f(2,800,2,t-epsi))/(2*epsi)

def dx_adelante(f,t):
    return (f(2,800,2,t+epsi)-f(2,800,2,t))/epsi
    
```

Metodo de biseccion usando la Derivada Exacta tomando a = -0.36 y b = -0.15

```python
def biseccion(f,a,b,n):
    aprox = num.zeros(n)
    for i in range(0,n):
        c = (a+b)/2
        if f(c)*f(a) > 0:
            a = c
        else: 
            b = c
        aprox[i] = c
    return c,aprox

n = 13  
a = -0.36
b = -0.15

plt.scatter(num.linspace(start=0,stop=n,num=n),biseccion(dx_exacta,a,b,n)[1])
plt.axhline(y = -0.23562,color='r')

        
    

```
